# Exploratory Data Analysis

```python
df.columns
```

**Gender**

```python
ax = sns.countplot(data=df, x="Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_pow = df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending=False)

sns.barplot(x = "Gender", y= "Amount", data = sales_pow)
```

*from above graphs, we can see that most of the buyers are females and even the purchasing power of females are greater than man*

**Age**

```python
df.columns
```

```python
sns.countplot(data=df, x="Age Group")
```

```python
# hue --> it is used so to bifurcate the "Gender"

sns.countplot(data=df, x="Age Group", hue = "Gender")
```

```python
ax = sns.countplot(data=df, x="Age Group", hue = "Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.head()
```

```python
sales_age = df.groupby(["Age Group"], as_index = False)["Amount"].sum().sort_values(by = "Amount",ascending=False)

sns.barplot(data= sales_age, x = "Age Group", y = "Amount")
```

*From above graphs we can see that most of the buyers are from Age Group b/w 26-35 yrs female*

**State**

```python
df.columns
```

```python
# Total number of Orders from top 10 states

sales_state = df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by = "Orders", ascending=False).head(10)

sns.barplot(data = sales_state, x = "State", y = "Orders")
```

```python
# Total number of Orders from top 10 states

# lets add figsize to look clearer

sales_state = df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by = "Orders", ascending=False).head(10)

plt.figure(figsize=(17,5))
sns.barplot(data = sales_state, x = "State", y = "Orders")
```

```python
# Total amount/sales from top 10 states

sales_amt = df.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending=False).head(10)

sns.barplot(data = sales_amt, x = "State", y = "Amount")
```

*From the above graph we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

**Marital Status**

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = "Marital_Status")

plt.figure(figsize=(4,3))

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marry = df.groupby(["Marital_Status", "Gender"], as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending=False)

sns.barplot(data= sales_marry, x = "Marital_Status", y = "Amount", hue = "Gender")
```

*From above graph we can see that most of the buyers are unmarried(womens) and they have high purchasing power*

**Occupation**

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = "Occupation")

plt.figure(figsize=(20,7))

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation = df.groupby(["Occupation"], as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending=False)

sns.barplot(data=sales_occupation, x = "Occupation", y = "Amount")
```

*From above graphs we can see that most of the buyers are working in IT, Aviation and Healthcare sector*

**Product Category**

```python
df.columns
```

```python

ax = sns.countplot(data = df, x = "Product_Category")

plt.figure(figsize=(25,10))

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(25,10))

sales_category = df.groupby(["Product_Category"], as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending=False).head(10)

sns.barplot(data=sales_category, x = "Product_Category", y = "Amount")
```

*From above graphs we can see that most of the sold products are from Food,Clothing & Electronics category*

**Top Products sold**

```python
df.columns
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
# Top 10 most sold products (same as above but with diff code)

fig1, ax1 = plt.subplots(figsize=(20,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).head(10).plot(kind='bar')
```
